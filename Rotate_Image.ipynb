```python
def load_Image(folder_nm,rotate_angle):
    from PIL import Image
    import os.path
    
    print("******************************","{}_{}".format(folder_nm,rotate_angle),"******************************")
    
    targerdir = r'C:/Users/soyeon/Desktop/soyeon_etc/computer_vision_code/실습/Plant_Disease/vision/training_set/{}/'.format(folder_nm) #이미지를 불러오는 경로 
    make_dir = r'C:/Users/soyeon/Desktop/soyeon_etc/computer_vision_code/실습/Plant_Disease/vision/training_set/' 
    files = os.listdir(targerdir) #이미지를 저장할 폴더 생성
    format = [".jpg",".png",".jpeg","bmp",".JPG",".PNG","JPEG","BMP"] #이미지 포맷
    
    angle =rotate_angle
    os.makedirs(make_dir+"{}_{}".format(folder_nm,rotate_angle), exist_ok=True)
    
    for (path,dirs,files) in os.walk(targerdir):
        for file in files:
            if file.endswith(tuple(format)):
                image = Image.open(path+file)
                print(image.filename)
                
                image=image.rotate(angle)
                image.save(make_dir+"{}_{}".format(folder_nm,rotate_angle)+"/"+"{}_{}_".format(folder_nm,rotate_angle)+file)
                
            else:
                print(path)

    
```

```python
load_Image('burn_disease',90)
load_Image('burn_disease',180)
```

```python
load_Image('healthy',90)
load_Image('healthy',180)
load_Image('healthy',270)
```
